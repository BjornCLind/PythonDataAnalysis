# About the Dataset

## Context
This dataset comes from a survey conducted among secondary school students who take math and Portuguese language courses. It includes a wealth of social, gender, and study-related information about the students. The data can be used for exploratory data analysis (EDA) or even to build predictive models to forecast students’ final grades.

## Content
The dataset is available in two files:
- **student-mat.csv** (Math course)
- **student-por.csv** (Portuguese language course)

Each dataset contains the following attributes:

- **school:** Student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
- **sex:** Student's sex (binary: 'F' - female or 'M' - male)
- **age:** Student's age (numeric: from 15 to 22)
- **address:** Student's home address type (binary: 'U' - urban or 'R' - rural)
- **famsize:** Family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
- **Pstatus:** Parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
- **Medu:** Mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education, 4 – higher education)
- **Fedu:** Father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education, 4 – higher education)
- **Mjob:** Mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g., administrative or police), 'at_home' or 'other')
- **Fjob:** Father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g., administrative or police), 'at_home' or 'other')
- **reason:** Reason to choose the school (nominal: close to 'home', school 'reputation', 'course' preference, or 'other')
- **guardian:** Student's guardian (nominal: 'mother', 'father' or 'other')
- **traveltime:** Home-to-school travel time (numeric: 1 - <15 min, 2 - 15 to 30 min, 3 - 30 min to 1 hour, or 4 - >1 hour)
- **studytime:** Weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
- **failures:** Number of past class failures (numeric: n if 1 ≤ n < 3, else 4)
- **schoolsup:** Extra educational support (binary: yes or no)
- **famsup:** Family educational support (binary: yes or no)
- **paid:** Extra paid classes within the course subject (binary: yes or no)
- **activities:** Extra-curricular activities (binary: yes or no)
- **nursery:** Attended nursery school (binary: yes or no)
- **higher:** Wants to take higher education (binary: yes or no)
- **internet:** Internet access at home (binary: yes or no)
- **romantic:** With a romantic relationship (binary: yes or no)
- **famrel:** Quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
- **freetime:** Free time after school (numeric: from 1 - very low to 5 - very high)
- **goout:** Frequency of going out with friends (numeric: from 1 - very low to 5 - very high)
- **Dalc:** Workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
- **Walc:** Weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
- **health:** Current health status (numeric: from 1 - very bad to 5 - very good)
- **absences:** Number of school absences (numeric: from 0 to 93)

### Grades Related to the Course Subject
The following columns represent the students’ grades:
- **G1:** First period grade (numeric: from 0 to 20)
- **G2:** Second period grade (numeric: from 0 to 20)
- **G3:** Final grade (numeric: from 0 to 20, and serves as the output target for predictive tasks)

### Additional Notes
- **Overlap Between Datasets:**  
  There are 382 students that belong to both the math and Portuguese language datasets. These students can be identified by searching for identical attributes that characterize each student.

## Assessing Academic Performance

In this section, we aim to understand overall student performance by examining the distribution of the grades and analyzing how final grades (G3) relate to the earlier period grades (G1 and G2).

### 1. Distribution of Grades

We start by exploring the distribution of the three grade variables:
- **G1:** First period grade
- **G2:** Second period grade
- **G3:** Final grade (target variable)

Visualizing these distributions using histograms can help us detect patterns such as skewness, clustering, or any anomalies.

````python
# Scatter plot for G1 vs. G3
plt.figure(figsize=(8,6))
sns.scatterplot(x='G1', y='G3', data=df)
plt.title('G1 vs G3')
plt.xlabel('G1 (First Period Grade)')
plt.ylabel('G3 (Final Grade)')
plt.show()

# Scatter plot for G2 vs. G3
plt.figure(figsize=(8,6))
sns.scatterplot(x='G2', y='G3', data=df)
plt.title('G2 vs G3')
plt.xlabel('G2 (Second Period Grade)')
plt.ylabel('G3 (Final Grade)')
plt.show()

# grade_corr = df[['G1', 'G2', 'G3']].corr()
# print("Correlation matrix for grades:")
# print(grade_corr)

````

### Interpretation of the Grades Correlation Matrix

The correlation matrix for the grade variables (G1, G2, G3) is as follows:

|     |   G1    |   G2    |   G3    |
|-----|---------|---------|---------|
| **G1** | 1.000000 | 0.852118 | 0.801468 |
| **G2** | 0.852118 | 1.000000 | 0.904868 |
| **G3** | 0.801468 | 0.904868 | 1.000000 |

#### Key Observations:
- **Strong Correlation Between Early Grades:**  
  The correlation between G1 (first period) and G2 (second period) is 0.85, indicating a very strong positive relationship. This suggests that students who perform well in the first period tend to also perform well in the second period.
  
- **G2 as a Strong Predictor of Final Grade (G3):**  
  The correlation between G2 and G3 is 0.90, which is even higher. This implies that the second period grade is an excellent indicator of the final grade.
  
- **G1 and Final Grade (G3):**  
  The correlation between G1 and G3 is 0.80, which is still strong but slightly lower than the G2-G3 correlation. This might suggest that the performance in the first period, while important, is somewhat less predictive of the final outcome than the second period grade.

#### Implications:
- **Consistency in Performance:**  
  The high correlations indicate a strong consistency in student performance over the course of the year. Early grades are good predictors of the final outcome.
  
- **Potential for Early Intervention:**  
  Given that G2 is highly correlated with G3, educators could focus on the second period performance as a key metric for identifying students who might need additional support before the final grade is determined.

Overall, the correlations suggest that while all three grading periods are closely linked, the second period grade (G2) may provide the most reliable signal for the final academic performance (G3).

````python
# List all categorical columns (excluding 'Medu_cat'/'Fedu_cat' if you created them)
categorical_cols = [
    'school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 
    'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 
    'nursery', 'higher', 'internet', 'romantic'
]
# Calculate and display mean G3 for each category
for cat in categorical_cols:
    print(f"\n--- {cat} ---")
    mean_scores = df.groupby(cat)['G3'].mean().sort_values(ascending=False)
    print(mean_scores)
    
    # Optional: create a boxplot or bar chart
    plt.figure(figsize=(6,4))
    sns.boxplot(x=cat, y='G3', data=df, order=mean_scores.index)
    plt.title(f"Final Grade (G3) Distribution by {cat}")
    plt.xlabel(cat)
    plt.ylabel("Final Grade (G3)")
    plt.tight_layout()
    plt.show()

    # Example: Combination of 'sex' and 'school'
combo_scores = df.groupby(['sex', 'school'])['G3'].mean().sort_values(ascending=False)
print("\nMean G3 by Sex and School:")
print(combo_scores)

# Convert groupby result to a pivot table for a heatmap
pivot_combo = df.pivot_table(values='G3', index='sex', columns='school', aggfunc='mean')
sns.heatmap(pivot_combo, annot=True, cmap='viridis')
plt.title("Mean Final Grade (G3) by Sex and School")
plt.show()
````

````python
from scipy.stats import f_oneway

# Example: ANOVA for 'Mjob' categories
groups = []
for job_category in df['Mjob'].unique():
    # G3 scores for each Mjob category
    groups.append(df[df['Mjob'] == job_category]['G3'].dropna())

f_stat, p_val = f_oneway(*groups)
print("ANOVA results for Mjob and G3:")
print("F-statistic:", f_stat)
print("p-value:", p_val)
````
